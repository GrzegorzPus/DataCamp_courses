### The urllib package

* Provides interface for fetching data across the web

* <code>urlopen()</code> - accepts URLs instead of file names

````python
from urllib.request import urlretrieve
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
urlretrieve(url, 'winequality-white.csv')
````

The `urlretrieve` function is part of the Python `urllib.request` module, and it is used to retrieve a file from a specified URL and save it to a local file on your computer. Here's a breakdown of its usage and parameters:

```python
urllib.request.urlretrieve(url, filename=None, reporthook=None, data=None)
```

- `url` (string): The URL of the file you want to retrieve.
- `filename` (string, optional): The local filename where the retrieved file should be saved. If not specified, the function will attempt to determine the filename from the URL.
- `reporthook` (callable, optional): A function that can be used to report the download's progress. It is called with a block number, a read size, and the total file size. This parameter is often used to display download progress in a user-friendly way.
- `data` (bytes or None, optional): Data to be sent with the request. This is typically used when making POST requests.

The function downloads the file specified by the URL and saves it to the local file with the specified filename. If no filename is provided, it will try to determine the filename from the URL. If `reporthook` is specified, it can be used to monitor the download progress.

In your example, `urlretrieve(url, 'winequality-red.csv')` is used to retrieve a file from the given URL (`url`) and save it locally as 'winequality-red.csv'.

### HTTP requests to import files from the web

**URL**
* Uniform/Universal Resource Locator
* References to web resources
* Focus: web addresses
*Ingredients:
** Protocol identifier - http:
** Resource name - datacamp.com
* These specify web addresses uniquely


````python
# Using urllib
from urllib.request import urlopen, Request
url = "https://www.wikipedia.org/"

request = Request(url)

response = urlopen(request)

html = response.read()
response.close()
````

This code demonstrates how to use the `urllib` library in Python to perform a simple HTTP GET request to fetch the HTML content of a web page. Here's a breakdown of each part of the code:

1. Importing Required Modules:
   ```python
   from urllib.request import urlopen, Request
   ```
   This line imports two classes, `urlopen` and `Request`, from the `urllib.request` module. These classes are used for making HTTP requests.

2. Defining the URL:
   ```python
   url = "https://www.wikipedia.org/"
   ```
   The URL of the web page you want to fetch is defined here. In this case, it's set to the Wikipedia homepage.

3. Creating a Request Object:
   ```python
   request = Request(url)
   ```
   A `Request` object is created using the URL. This object represents the HTTP request that will be sent to the server.

4. Sending the HTTP Request:
   ```python
   response = urlopen(request)
   ```
   The `urlopen` function is used to send the HTTP GET request specified by the `request` object. It sends the request to the server and receives the server's response. The response is stored in the `response` variable.

5. Reading the Response:
   ```python
   html = response.read()
   ```
   The `read()` method is called on the `response` object to read the content of the web page. The HTML content is stored in the `html` variable as bytes.

6. Closing the Response:
   ```python
   response.close()
   ```
   It's important to close the response object when you're done with it to release system resources.

Now, the `html` variable contains the HTML content of the web page specified by the URL. You can then process or parse this HTML content for further use, such as scraping data from the page or performing other actions.

````python
# Import packages
import requests
from bs4 import BeautifulSoup

# Specify url: url
url = 'https://www.python.org/~guido/'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extract the response as html: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc, 'html.parser')

# Get the title of Guido's webpage: guido_title
guido_title = soup.title

# Print the title of Guido's webpage to the shell
print(guido_title)

# Get Guido's text: guido_text
guido_text = soup.get_text()

# Print Guido's text to the shell
print(guido_text)
````
